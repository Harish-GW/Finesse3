```python
import finesse
finesse.configure(plotting=True)
le=finesse.Model()
le.parse("""
l l1 P=10
s s0 l1.p1 md.p1 L=0
mod md midx=0.5 order=1 f=0.75e6 
s s1 md.p2 m1.p1 L=0
m m1 R=0.9 T=0.1
s cav m1.p2 m2.p1 L=2
m m2 R=1 T=0


#ad Circulated_c m1.p2.o f=0
#ad Circulated_s m1.p2.o f=0.75e7
pd Circulated m1.p2.o
pd1 Error_i m1.p1.o f=0.75e6 phase=0
pd1 Error_q m1.p1.o f=0.75e6 phase=90
xaxis(m2.phi,lin,-180,180,10000)
""")
```

```python

out=le.run()
out.plot()
```
