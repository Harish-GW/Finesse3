```python
import finesse
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig
```

```python
#Impedance matching the FPC
fpc=finesse.Model()
finesse.configure(plotting=True)
fpc.parse(f"""
var R 0
l l1 P=1
s s0 l1.p1 m1.p1 L=0
m m1 R=R L=8.5e-6 T=1-8.5e-6-R
s cav m1.p2 m2.p1 L=10
m m2 R=0.9 L=8.5e-6 

pd Circ m1.p2.o
xaxis(R,lin,0,1,100000)

""")
out=fpc.run()

Circ=[]
for i in range(100000):
    Circ.append(out["Circ"][i])
    
#Finding R, ideally must be equal to 0.9 (impedance matched)
R=np.linspace(0,1,100000)
R_opt,_=sig.find_peaks(Circ,R)

plt.plot(R,Circ)

print(R[R_opt])
```

```python
fpc.m1.R=R[R_opt]
fpc.m1.T=1-8.5e-6-R[R_opt]
fpc.parse("xaxis(m2.phi,lin,-180,180,100000)")

out=fpc.run()
phi=np.linspace(-180,180,100000)
Circ=[]
for i in range(100000):
    Circ.append(out["Circ"][i])
plt.plot(phi,Circ)
```
