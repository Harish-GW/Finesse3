#### Adding random PRC length parameters (Lx,Ly,pr-bs)

```python
prmi3=finesse.Model()
prmi3.parse(f"""
l laser P=1
s s0 laser.p1 eom.p1 L=0
mod eom midx=0.4 order=1 f=9M

#PRM
s s1 eom.p2 pr.p1 L=0
m pr R=0.85 T=0.15           
s pr_Bs pr.p2 Bs.p1 L=20  

bs Bs R=0.5 T=0.5 

#X-arm
s Lx Bs.p3 ITMx.p1 L=5
m ITMx R=0.9 T=0.1 phi=90

#Y-arm
s Ly Bs.p2 ITMy.p1 L=5
m ITMy R=0.9 T=0.1

#pd AS_dc Bs.p4.o 
#pd REFL_dc pr.p1.o
#pd POP_dc pr.p2.o

ad POP_C pr.p2.o f=0
ad POP_9M pr.p2.o f=9M
#ad AS_C Bs.p4.o f=0
#ad AS_9M Bs.p4.o f=9M

xaxis(pr.phi,lin,-180,180,1000)

""")
output=prmi3.run()
output.plot()
```

#### Optimising the random value by varying the pr-bs length.

```python
import numpy as np
phi=np.linspace(0,180,500)
lengths=np.linspace(19,22,500)
possible_pr_bs_error=[]

for l in np.linspace(19,22,500):
    prmi3.pr_Bs.L=l
    output=prmi3.run()
    POP_9M=[]
    for i in range(500,1000):
        POP_9M.append(abs(output["POP_9M"][i]))
    possible_pr_bs_error.append(abs(phi[POP_9M.index(max(POP_9M))]-90)) #90 because the anti-resonance of the carrier occurs there.
                                                                        #(peak occurs every 180 deg)      
    
pr_bs_l_opt1=lengths[possible_pr_bs_error.index(min(possible_pr_bs_error))]
print(pr_bs_l_opt1)
```

#### In the above cells, I have varied the the pr-bs length from 3.2 to 3.4. I did that to find the optimum pr-bs length that will see res in C and anit-res in S together.
#### I cannot find the exact value using python, I will have to go back modelling the field equations and see for which pr-bs length, they occur together. So I did a search by changing the length values 3000 times! and finding the pr.phi at which SB is resonant. I already knew that the Carrier anti-resonance occurs at pr.phi=90 degrees (irrespective of the pr-bs length). 
#### So I calculated the error (difference between 90 and the pr.phi at which S is res) for all these 3000 iterations. Then I found the minimum of this error and set the corresponding length to be the optimum pr-bs length :) 
#### I stopped at 3.4, because my initial guess was 3.35m (with naked eye).

#### I am setting this optimum lenth.

```python
prmi3=finesse.Model()
prmi3.parse(f"""
l laser P=1
s s0 laser.p1 eom.p1 L=0
mod eom midx=0.4 order=1 f=9M

#PRM
s s1 eom.p2 pr.p1 L=0
m pr R=0.85 T=0.15           
s pr_Bs pr.p2 Bs.p1 L={pr_bs_l_opt1}  

bs Bs R=0.5 T=0.5 

#X-arm
s Lx Bs.p3 ITMx.p1 L=5
m ITMx R=0.9 T=0.1 phi=90

#Y-arm
s Ly Bs.p2 ITMy.p1 L=5
m ITMy R=0.9 T=0.1

#pd AS_dc Bs.p4.o 
#pd REFL_dc pr.p1.o
#pd POP_dc pr.p2.o

ad POP_C pr.p2.o f=0
ad POP_9M pr.p2.o f=9M
#ad AS_C Bs.p4.o f=0
#ad AS_9M Bs.p4.o f=9M

xaxis(pr.phi,lin,-180,180,1000)

""")
output1=prmi3.run()
output1.plot()
```

#### To bring the necessary resonances to zero phase of xaxis, we will need to set the tuning phase.
#### For which, we will need to know the pr.phi at which the sideband attains a resonance, ideally it must be the angle I must give to bring the anti-resonance of carrier, which is 90 deg. But since the PRC length parameters are chosen randomly, these two do not match!
#### I am finding the pr.phi at which sideband gets resonance and tuning the PRM by that amount.
#### I am setting this angle and not 90 degrees because, I believe the Carrier need not be in a proper anti-resonance, It can be in a very close vicinity of it.

```python
Amplitude=[]
pr_phi=np.linspace(0,180,500) #avoiding the peak when pr.phi is negative at around -90 on the previous plot.
for i in range(500,1000,1):
    Amplitude.append(abs(output1["POP_9M"][i]))
max_amp=max(Amplitude)
pr_phi_opt1=pr_phi[Amplitude.index(max_amp)]
print("pr_phi_opt:",pr_phi_opt1) #It is slightly more than 90 degrees.
```

#### I am setting this optimum pr.phi.

```python
prmi5=finesse.Model()
prmi5.parse(f"""
l laser P=1
s s0 laser.p1 eom.p1 L=0
mod eom midx=0.4 order=1 f=9M

#PRM
s s1 eom.p2 pr.p1 L=0
m pr R=0.85 T=0.15 phi={pr_phi_opt1}          
s pr_Bs pr.p2 Bs.p1 L={pr_bs_l_opt1}  

bs Bs R=0.5 T=0.5 

#X-arm
s Lx Bs.p3 ITMx.p1 L=5
m ITMx R=0.9 T=0.1 phi=90

#Y-arm
s Ly Bs.p2 ITMy.p1 L=5
m ITMy R=0.9 T=0.1

pd AS_dc Bs.p4.o 
#pd REFL_dc pr.p1.o
#pd POP_dc pr.p2.o

ad POP_C pr.p2.o f=0
ad POP_9M pr.p2.o f=9M
#ad AS_C Bs.p4.o f=0
#ad AS_9M Bs.p4.o f=9M

xaxis(Bs.phi,lin,-180,180,1000)
""")
output=prmi5.run()
output.plot()
```
