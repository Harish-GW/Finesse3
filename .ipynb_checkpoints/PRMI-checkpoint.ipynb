```python
import finesse
finesse.configure(plotting=True)
sm=finesse.Model()
sm.parse("""
l laser1 P=1
s s1 laser1.p1 pr.p1 L=0
m pr R=0.9 T=0.1     #Power recylcing mirror
s s2 pr.p2 bs1.p1 L=1
bs bs1 R=0.5 T=0.5   #50-50 Beam Splitter
s Lx bs1.p3 ETMx.p1 L=1+2.087564288E-6 #Impedance matched(R_pr=sin^2(k(lx-ly)))
m ETMx R=1 T=0 phi=90                    #Setting the asymmeteric port dark

s Ly bs1.p2 ETMy.p1 L=1
m ETMy R=1 T=0 
pd Circulated bs1.p1.i      #Power circulated in the Power recycling cavity
pd Transmitted bs1.p4.o     #Power trasnmitted out to the asymmertric port
pd Relfected pr.p1.o        #Power reflected from the PR mirror
xaxis(laser1.f,lin,1E8,2E8,10000)
""")
```

```python
out=sm.run()
out.plot()
circulated=[]
for i in range(10001):
    circulated.append(out["Circulated"][i])
```

```python
b=[]
for i in range(10001):
    if abs(circulated[i]-0.5*max(circulated))<0.15:#approximating the FWHM by finding two points close to the half maximum
        b.append(circulated[i])
print(b)
```

```python
print("FWHM_pr={}Hz".format(1E4*(circulated.index(b[1])-circulated.index(b[0]))))
print("FSR_pr={}Hz".format(1.5E8))
print("FINESSE_pr={}".format(1.5E8/(1E4*(circulated.index(b[1])-circulated.index(b[0])))))
```
