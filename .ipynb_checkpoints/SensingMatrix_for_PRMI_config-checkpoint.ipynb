```python
import finesse
finesse.configure(plotting=True)
sm=finesse.Model()
sm.parse("""
l laser P=1 

s s0 laser.p1 md1.p1 L=0
mod md1 midx=0.4 order=1 f=0.75e8

s s2 md1.p2 pr.p1 L=0
m pr R=0.9 T=0.1 phi=90          
s PRC pr.p2 bs1.p1 L=2
bs bs1 R=0.5 T=0.5 

s Lx bs1.p3 ITMx.p1 L=0.99
m ITMx R=1 T=0 phi=90

s Ly bs1.p2 ITMy.p1 L=1.01
m ITMy R=1 T=0 

ad POP_c pr.p2.o f=0
ad POP_s pr.p2.o f=0.75e8
ad AS_c bs1.p4.o f=0
ad AS_s bs1.p4.o f=0.75e8

xaxis(laser.f,lin,-0.25e8,0.25e8,10000)        
""")
```

```python
out=sm.run()
out.plot()
```
