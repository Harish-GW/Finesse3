```python
import finesse
finesse.configure(plotting=True)
le=finesse.Model()
le.parse("""
l l1 P=1 
s s0 l1.p1 md.p1 L=0
mod md midx=0.4 order=1 f=0.74e7   
s s1 md.p2 m1.p1 L=0
m m1 R=0.85 T=0.15
s cav m1.p2 m2.p1 L=1
m m2 R=1 T=0


#ad Circulated_c m1.p2.o f=0
#ad Circulated_s m1.p2.o f=md.f
#ad Reflected_c m1.p1.o f=0
pd Circulated m1.p2.o
pd1 Error_i m1.p1.o f=md.f phase=0  #I-phase Error
#pd1 Error_q m1.p1.o f=md.f phase=90 #Q-phase Error
xaxis(l1.f,lin,-1.5e8,1.5e8,10000)
""")
```

```python
out=le.run()
out.plot(separate=True)
```

```python
le.md.f=0.75e8
le.Error_i.f=le.md.f
le.run().plot()
```
